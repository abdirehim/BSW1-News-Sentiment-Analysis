```python
# 1. Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load data
sentiment_df = pd.read_csv("../output/daily_sentiment.csv")
returns_df = pd.read_csv("../output/daily_returns.csv")

# 3. Merge on ['stock', 'publish_day']
merged = pd.merge(sentiment_df, returns_df, on=["stock", "publish_day"])
print(merged.head())

# 4. Correlation
print("Pearson correlation:", merged["avg_sentiment"].corr(merged["daily_return"]))
```

```python
# 5. Optional: Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=merged, x="avg_sentiment", y="daily_return", hue="stock", alpha=0.7)
plt.title("Sentiment vs Daily Return")
plt.grid(True)
plt.show()
```
